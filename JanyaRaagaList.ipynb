```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def fetch_and_parse_url(url):
    response = requests.get(url)
    return BeautifulSoup(response.content, 'html.parser')

def extract_table_data(table):
    rows = table.find_all('tr')
    table_data = []
    headers = [header.text.strip() for header in rows[0].find_all('th')]
    for row in rows[1:]:  # Skip the header row
        columns = row.find_all('td')
        data = [col.text.strip() for col in columns]
        table_data.append(data)
    return headers, table_data

def search_data(dataframe, search_string):
    return dataframe[dataframe.apply(lambda row: row.astype(str).str.contains(search_string, case=False).any(), axis=1)]

def main():
    url = 'https://en.wikipedia.org/wiki/List_of_Janya_ragas'
    soup = fetch_and_parse_url(url)

    all_data = []
    tables = soup.find_all('table', {'class': 'wikitable'})
    for table in tables:
        headers, data = extract_table_data(table)
        all_data.extend(data)

    df = pd.DataFrame(all_data, columns=headers)

    return df

df = main()
```

```python
while True:
        search_query = input("Enter search string (or 'exit' to quit): ")
        if search_query.lower() == 'exit':
            break
        filtered_data = search_data(df, search_query)
        print(filtered_data)
```
